```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```python
img = cv2.imread('/Users/amikar/Downloads/tmp878655630959706114.jpg',cv2.COLOR_BGR2GRAY)
blur = cv2.blur(img,(5,5))

res = cv2.resize(blur,None,fx=0.1, fy=0.1, interpolation = cv2.INTER_CUBIC)
surf = cv2.xfeatures2d.SURF_create(2000)
kp, des = surf.detectAndCompute(blur,None)
img2 = cv2.drawKeypoints(blur,kp,None,(255,0,0),4)
plt.imshow(img2)
plt.show()
print(len(kp))
```

```python
def rgb_to_hex(value1, value2, value3):

    for value in (value1, value2, value3):
        if not 0 <= value <= 255:
            raise ValueError('Value each slider must be ranges from 0 to 255')
    return "#{0:02X}{1:02X}{2:02X}".format(value1, value2, value3)
```

```python
average_color_per_row = np.average(res, axis=0)
```

```python
average_color = np.average(average_color_per_row, axis=0)
average_color = np.uint8(average_color)
print(average_color)
rgb_to_hex(average_color[0], average_color[1], average_color[2])
```
